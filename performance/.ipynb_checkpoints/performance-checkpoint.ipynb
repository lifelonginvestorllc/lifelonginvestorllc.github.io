```python
import pandas as pd
import numpy as np
import datetime as dt
from pandas_datareader import data as pdr
import yfinance as yf
yf.pdr_override()
```

```python
LLNAVDaily_df = pd.read_csv("LLNAVDaily.csv", header=None, skip_blank_lines=True, skiprows=2)
```

```python
LLNAVDaily_df.head()
```

```python
NAV = LLNAVDaily_df[LLNAVDaily_df.iloc[:,0]=="Allocation by Asset Class"].iloc[1:,:]
NAV = NAV.rename(columns=NAV.iloc[0]).drop(NAV.index[0]).dropna(axis=1)[['Date', 'NAV']]
NAV
```

```python
bench = LLNAVDaily_df[LLNAVDaily_df.iloc[:,0]=="Time Period Benchmark Comparison"].iloc[1:,:]
bench = bench.rename(columns=bench.iloc[0]).drop(bench.index[0]).dropna(axis=1)
format0 = '%m/%d/%Y'
bench['Date'] = pd.to_datetime(bench['Date']).dt.strftime(format0)
bench.set_index("Date", inplace=True)
```

```python
format0 = '%m/%d/%Y'
NAV['Date'] = pd.to_datetime(NAV['Date']).dt.strftime(format0)
NAV.set_index("Date", inplace=True)
```

```python
pd.concat([NAV, bench], axis=1)
```
